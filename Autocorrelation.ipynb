```python
def compute_mean(mass, lag, parametr):
    n = lag
    mass_len = len(mass) 
    mean = 0
    for i in range (n, mass_len):
        if (parametr == 't'):
            mean += mass[i]
        if (parametr == 't-1'):
            mean += mass[i - n]  
    mean = mean / (mass_len - n)
    return mean        
```

```python
def compute_autocorellation(mass, lag):
    numerator = 0
    denominator = 0
    denominator1 = 0
    denominator2 = 0
    for i in range (lag , len(mass)):
        numerator += (mass[i] - compute_mean(mass, lag, 't')) * (mass[i - lag] - compute_mean(mass, lag, 't-1'))
        denominator1 += (mass[i] - compute_mean(mass, lag, 't')) ** 2
        denominator2 += (mass[i - lag] - compute_mean(mass, lag, 't-1')) ** 2
    denominator = (denominator1 * denominator2) ** (0.5) 
    coeff = numerator / denominator
    return coeff
```

```python
for i in range (1,10):
    print(f'длина лага: {i}, коэффициент автокорелляции: {compute_autocorellation(data, i)}')
```
