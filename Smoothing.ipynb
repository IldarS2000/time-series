```python
import numpy as np
import pandas as pd

from pandas import read_excel
```

```python
df = read_excel('data.xlsx', sheet_name='2013-2018', engine='openpyxl')
df.head()
```

```python
data = list(df.iloc[0][1:])
print(data)
```

```python
n = len(data)
print(f'Объём выборки: {n}')
```

Метод простой скользящей средней

```python
smoothed_data = data.copy()
m = int(input('Введите значение m: '))

if m % 2 == 1:
    p = int((m - 1) / 2)
    for i in range(p, n - p):
        s = 0
        for j in range(i - p, i + p + 1):
            s = s + data[j]
        s = s / m
        smoothed_data[i] = s
else:
    p = int(m / 2)
    for i in range(p, n - p):
        s = data[i - p] / 2 + data[i + p] / 2
        for j in range(i - p + 1, i + p - 1 + 1):
            s = s + data[j]
        s = s / m
        smoothed_data[i] = s
```

```python
print(smoothed_data)
```

Метод экспоненциального сглаживания

```python
x = [0 for i in range(n + 1)]
x[0] = data[0] 

alpha = 2 / (n + 1)
print(f'alpha: {alpha}')

for i in range(0, n):
    x[i + 1] = alpha * data[i] + (1 - alpha) * x[i]  
```

```python
print(x) 
```

```python
forecast = x.pop(len(x) - 1)
print(f'Прогноз: {forecast}') 
```
